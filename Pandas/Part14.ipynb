```python
import numpy as np
import pandas as pd
```

```python
matches = pd.read_csv('matches - matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
# we can merge both the datasets
#delivery - left
#matches - right
new_data = delivery.merge(matches, left_on='match_id', right_on='id', how='inner')
new_data
```

```python
print(matches.shape)
print(delivery.shape)
```

```python
batsman = new_data.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first')
batsman.sort_values(by='season', ascending=False)
```
