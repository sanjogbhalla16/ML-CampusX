```python
def runs_scored(batsman_name):
    vk = data[data['batsman'] == batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
#we will first find the death overs
death_overs = data[data['over'] > 15]
```

```python
death_overs
```

```python
#now for each player how many balls he has faced in death overs
no_of_balls_in_death_overs_for_all_players = death_overs.groupby('batsman')['batsman'].count()
```

```python
#this will tell only those players who have faced more than 200 balls in death overs
x = no_of_balls_in_death_overs_for_all_players>200
```

```python
no_of_balls_in_death_overs_for_all_players[x]
```

```python
no_of_balls_in_death_overs_for_all_players[x].size
```

```python
batsman_list = no_of_balls_in_death_overs_for_all_players[x].index.tolist()
```

```python
#final = data[data['batsman'].isin(batsman_list)]
final = death_overs[death_overs['batsman'].isin(batsman_list)] 
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman'].count()
```

```python
strike_rate = (runs/balls)*100
```
